# python连接mysql数据库

```python
# 导包
import pymysql 

# 创建连接对象
mydb = pymysql.connect(  
                   host="localhost",          # 数据库主机地址
                   user="root",               # 数据库用户名
                   password="1234"  ,         # 数据库密码
                   database=None,             # 可以指定连接某个数据库
                   port=3306,                 # 端口号, 默认是3306
                   charset='utf8'             # 使用的编码
                   )  

# 创建游标对象, 这里创建的游标是  可以返回带字段名的字典， 比较好用
mycursor = mydb.cursor(cursor=pymysql.cursors.DictCursor)
```

# SQL

## 库操作

### 查看数据库

```python
# 通过游标对象的execute方法执行sql语句
mycursor.execute('''
                 show databases;
''')
pd.DataFrame(mycursor)
```

```python
mycursor.description  # 列名是'Database

# 这里也可以查询出列名
```

### 创建数据库  
新创建的数据库名字不能和sql中的关键字一样， 也不能和已经存在的数据库重名

```python
mycursor.execute('''
                 create database 数据库;
''')
pd.DataFrame(mycursor)  # 创建数据库不会有返回结果
```

```python
# 再次查询以便就有了  数据库
mycursor.execute('''
                 show databases;
''')
pd.DataFrame(mycursor)
```

### 删除数据库

```python
mycursor.execute('''
                 drop database 数据库;
''')
pd.DataFrame(mycursor)
```

### 查看已经创建好的数据库  
返回该数据的一些信息

```python
mycursor.execute('''
                 show create database 学生管理系统;
''')
a = pd.DataFrame(mycursor)
a
```

```python
a.values  # 可以看出该数据库编码是utf8
```

### 修改数据库编码

```python
mycursor.execute('''
                alter database 学生管理系统 character set utf8;
''')
```

```python
pd.DataFrame(mycursor)
```

```python
mycursor.execute('''
                 show create database 学生管理系统;
''')
a = pd.DataFrame(mycursor)
a
```

```python
a.values  # 编码已经修改成utf8
```

### 使用某个数据库

```python
mycursor.execute('''
                 use 学生管理系统;
''')
pd.DataFrame(mycursor)
```

### 创建表

#### 创建数据表stu

```python
# 创建数据库school
mycursor.execute('''
                 create database school;
''')

# 使用该数据库
mycursor.execute('''
                   use school;
''')
```

```python
# 创建表stu
mycursor.execute('''
                 create table stu(
                 s_id varchar(10) primary key,  
                 s_name varchar(10),
                 s_barth date,
                 s_sex varchar(10)           
                 );
''')
pd.DataFrame(mycursor)
```

更多数据类型和约束：https://www.cnblogs.com/LUOyaXIONG/p/12603975.html

#### 向表stu中插入数据

```python
mycursor.execute('''
insert into stu values
('01' , '赵雷' , '1990-01-01' , '男'),
('02' , '钱电' , '1990-12-21' , '男'),
('03' , '孙风' , '1990-05-20' , '男'),
('04' , '李云' , '1990-08-06' , '男'),
('05' , '周梅' , '1991-12-01' , '女'),
('06' , '吴兰' , '1992-03-01' , '女'),
('07' , '郑竹' , '1992-04-21' , '女'),
('08' , '王菊' , '1990-01-20' , '女');
''')
pd.DataFrame(mycursor)
```

```python
# 简单查询一下数据
mycursor.execute('''
                 select*from stu;
''')
stu = pd.DataFrame(mycursor)
stu
```

#### 查看当前数据库中所有表的表名

```python
mycursor.execute('''
                show tables;
''')
pd.DataFrame(mycursor)
```

#### 查看某张表的结构

```python
mycursor.execute('''
               desc stu;
''')
pd.DataFrame(mycursor)
```

#### 删除某张表

```python
mycursor.execute('''
                 drop table stu
''')
pd.DataFrame(mycursor)
```

```python
engine
```

```python
stu.to_sql(name='stu', con=engine, index=False)

# name是表名
# con是数据库引擎
# index=False忽略原行索引
```
