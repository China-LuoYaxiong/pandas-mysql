# python连接mysql数据库

```python
# 导包
import pymysql 

# 创建连接对象
mydb = pymysql.connect(  
                   host="localhost",          # 数据库主机地址
                   user="root",               # 数据库用户名
                   password="1234"  ,         # 数据库密码
                   database=None,             # 可以指定连接某个数据库
                   port=3306,                 # 端口号, 默认是3306
                   charset='utf8'             # 使用的编码
                   )  

# 创建游标对象, 这里创建的游标是  可以返回带字段名的字典， 比较好用
mycursor = mydb.cursor(cursor=pymysql.cursors.DictCursor)
```

```python
np.random.seed(999)
a = np.random.randint(60, 100, (10, 6))
df = pd.DataFrame(a)
df.columns = ['语文', '数学', '英语', '化学', '物理', '生物']
df.index = ['张三', '李四', '王五', '赵一', '张四', 
            '刘六', '张五', '李一', '周四', '赵四']
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'姓名'})
s = '男女男男男女男女男女'
df.insert(0, '性别', list(s))
df.insert(5, '三总', df.语文+df.数学+df.英语)
df['理综'] = df.物理+df.化学+df.生物
df.insert(0, '总分', df.三总+df.理综)
df
```

```python
# 创建数据库引擎
from sqlalchemy import create_engine
engine = create_engine("mysql+mysqlconnector://root:1234@localhost:3306/school?charset=utf8")

# 这里需要先安装：pip install mysql-connector-python -i https://pypi.douban.com/simple
# 这里如果使用pymysql会有一点问题， 建议就使用mysql-connector-python
```

```python
df.to_sql(name='score', con=engine, index=False, )

# name是表名
# con是数据库引擎
# index=False忽略原行索引
```

```python
mycursor.execute('''
                 select*from score;      
''')
pd.DataFrame(mycursor)
```

# 索引可切片相似操作

```python
mycursor.execute('''
                 select 姓名, 语文, 物理 from score;      
''')
pd.DataFrame(mycursor)
```

# 布尔索引相似操作-where语句

## 挑选出 三总 成绩最高的 前三名 同学的所以信息

```python
df.sort_values('三总', ascending=False)[0:3]
```

```python
mycursor.execute('''
                 select * from score
                 order by 三总 desc
                 limit 3
                 ;      
''')
pd.DataFrame(mycursor)
```

## 挑选出 物理 大于 80分的同学信息

```python
df[df.物理>80]
```

```python
mycursor.execute('''
select * from score
where 物理>80;
''')
pd.DataFrame(mycursor)
```

## 挑选出 数学大于70 并且 英语 大于80的同学信息

```python
df[((df.数学>70)&(df.英语>80))]
```

```python
mycursor.execute('''
select * from score
where 数学>70 and 英语>80
''')
pd.DataFrame(mycursor)
```

## 挑选所有男同学

```python
df[df.性别=='男']
```

```python
mycursor.execute('''
select * from score
where 性别='男'
''')
pd.DataFrame(mycursor)
```

## 查看女同学语文的平均成绩

```python
df[df.性别=='女'].语文.mean()
```

```python
mycursor.execute('''
select avg(语文) as 女同学语文平均成绩 from score
where 性别='女'
''')
pd.DataFrame(mycursor)
```

## 查看总分在480分以上的男同学的理综的平均成绩

```python
df[(df.总分>480)&(df.性别=='男')].理综.mean()
```

```python
mycursor.execute('''
select avg(理综)  from score
where 总分>480 and 性别='男';
''')
pd.DataFrame(mycursor)
```

## 查询数学成绩比张三至少高2分， 物理成绩比张三至少高5分的同学(有疑问)

```python
df
```

```python
df[df.姓名=='张三'].数学.values[0]
```

```python
df.数学>df[df.姓名=='张三'].数学.values[0]+2
```

```python
df.物理>df[df.姓名=='张三'].物理.values[0]+5
```

```python
df[(df.数学>df[df.姓名=='张三'].数学.values[0]+2)&(df.物理>df[df.姓名=='张三'].物理.values[0]+5)]
```

```python
mycursor.execute('''
select * from score
where (数学>(select 数学+2 from score where 姓名='张三'))
       and
       (物理>(select 物理+5 from score where 姓名='张三'))
''')
pd.DataFrame(mycursor)
```

```python
# 还可以
mycursor.execute('''
select * from score
where (数学, 物理)>(select 数学+2, 物理+5 from score where 姓名='张三')
''')
pd.DataFrame(mycursor)
```
